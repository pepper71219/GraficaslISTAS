```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,4,9,16,25]


plt.plot(x, y)
plt.show
plt.plot(x, y, label='x^2')
plt.title('Simple line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axix')
plt.show()
```
